### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data). 

### Step 3. Assign it to a variable called iris

```python
data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris_df = pd.read_csv(data)

iris_df[:5]
```

### Step 4. Create columns for the dataset

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class

cols = ['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']

iris_df.columns = cols

iris_df.head()
```

### Step 5.  Is there any missing value in the dataframe?

```python
iris_df[iris_df.isnull() == False]

iris_df.isnull().sum()

# iris_df.notnull() == False

# Ah, this is much better, use top-level function
# pd.isnull(iris_df).sum()

# Nvm, can do that on the instance-level function
```

### Step 6.  Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
# iris_df[9:30].petal_len = 'NaN'

# iris_df.loc[10:29].petal_len = 'NaN' Almost

iris_df.loc[10:30, 'petal_len'] = np.nan

# Needed numpy here

```

### Step 7. Good, now lets substitute the NaN values to 1.0

```python
iris_df[10:12]

# iris_df.isnull()

# Bingo

# iris_df.isnull = 1 Oh rofl whast I did here was set the method to equal 1 haha

iris_df[10:12]


# Needed to do column level

iris_df.petal_len.fillna(value=1, inplace=True)

iris_df[10:30]
```

### Step 8. Now let's delete the column class

```python
del iris_df['class']

iris_df.head()
```

### Step 9.  Set the first 3 rows as NaN

```python
iris_df.loc[0:2] = np.nan
```

### Step 10.  Delete the rows that have NaN

```python
iris_df[iris_df.notnull() == False]

# Ah! Dropna

iris_df.dropna(inplace=True)

iris_df.head()
```

### Step 11. Reset the index so it begins with 0 again

```python
iris_df.index.reindex(iris_df.index)

iris_df.reset_index(inplace=True, drop=True)

iris_df.head()
```
