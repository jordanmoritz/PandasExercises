### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'

drinks = pd.read_csv(url)
```

```python
drinks.head(5)

drinks.info()

drinks.describe(include='all')
```

### Step 4. Which continent drinks more beer on average?

```python
continent_beer = drinks[['beer_servings', 'continent']]

continent_beer.groupby('continent').mean()

# Same as

drinks[['beer_servings', 'continent']].groupby('continent').mean().sort_values(by='beer_servings', ascending=False)

# Same as

drinks.groupby('continent').beer_servings.mean().sort_values(ascending=False)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks[['wine_servings', 'continent']].groupby('continent').describe()
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
drinks.groupby(by='continent').mean()
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
drinks.groupby(by='continent').median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max']).sort_values(by='max', ascending=False)
```
